```python
### 1 st question. write a Python script that reads the students.
# jon JSON file and prints details of each student.
import json

def print_student_details():
    try:
        
        with open('students.json', 'r') as file:
            students = json.load(file)

        
        for i, student in enumerate(students, start=1):
            print(f"\nStudent {i}:")
            for key, value in student.items():
                print(f"  {key.capitalize()}: {value}")

    except FileNotFoundError:
        print("Error: 'students.json' file not found.")
    except json.JSONDecodeError:
        print("Error: Invalid JSON format in 'students.json'.")


if __name__ == "__main__":
    print_student_details()
```

```python
# 2nd question. 
import requests

def get_weather(city_name, api_key):
    url = "http://api.openweathermap.org/data/2.5/weather"
    params = {
        'q': city_name,
        'appid': api_key,
        'units': 'metric'  # 'metric' for Celsius, 'imperial' for Fahrenheit
    }

    try:
        response = requests.get(url, params=params)
        data = response.json()

        if response.status_code == 200:
            print(f"\nWeather in {city_name}:")
            print(f"  Temperature: {data['main']['temp']} ¬∞C")
            print(f"  Humidity   : {data['main']['humidity']} %")
            print(f"  Pressure   : {data['main']['pressure']} hPa")
            print(f"  Weather    : {data['weather'][0]['description'].capitalize()}")
            print(f"  Wind Speed : {data['wind']['speed']} m/s")
        else:
            print(f"Error {response.status_code}: {data.get('message', 'Unable to fetch data')}")

    except requests.exceptions.RequestException as e:
        print(f"Request error: {e}")

# Your API key and city
API_KEY = '817520a7ecc14e56c135f2cdbe1d59b0'
CITY = 'Tashkent'

# Run it
if __name__ == "__main__":
    get_weather(CITY, API_KEY)
```

```python
## 3 rd question. Task: JSON Modification
#Write a program that allows users to add new books, update existing book information, 
# and delete books from the books.json JSON file.
import json
import os
BOOKS_FILE = 'books.json'

def load_books():
    if not os.path.exists(BOOKS_FILE):
        return []
    with open(BOOKS_FILE, 'r') as file:
        return json.load(file)

def save_books(books):
    with open(BOOKS_FILE, 'w') as file:
        json.dump(books, file, indent=4)

def add_book():
    books = load_books()
    try:
        new_id = int(input("Enter book ID: "))
        if any(book["id"] == new_id for book in books):
            print("A book with that ID already exists.")
            return
        title = input("Enter book title: ")
        author = input("Enter book author: ")
        books.append({"id": new_id, "title": title, "author": author})
        save_books(books)
        print("Book added successfully.")
    except ValueError:
        print("Invalid input. ID must be a number.")

def update_book():
    books = load_books()
    try:
        book_id = int(input("Enter the ID of the book to update: "))
        for book in books:
            if book["id"] == book_id:
                book["title"] = input(f"New title (current: {book['title']}): ") or book["title"]
                book["author"] = input(f"New author (current: {book['author']}): ") or book["author"]
                save_books(books)
                print("Book updated successfully.")
                return
        print("Book not found.")
    except ValueError:
        print("Invalid input. ID must be a number.")

def delete_book():
    books = load_books()
    try:
        book_id = int(input("Enter the ID of the book to delete: "))
        new_books = [book for book in books if book["id"] != book_id]
        if len(new_books) == len(books):
            print("Book not found.")
        else:
            save_books(new_books)
            print("Book deleted successfully.")
    except ValueError:
        print("Invalid input. ID must be a number.")

def list_books():
    books = load_books()
    if not books:
        print("No books found.")
    else:
        for book in books:
            print(f"ID: {book['id']}, Title: {book['title']}, Author: {book['author']}")

def menu():
    while True:
        print("\nBook Manager Menu:")
        print("1. List books")
        print("2. Add book")
        print("3. Update book")
        print("4. Delete book")
        print("5. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            list_books()
        elif choice == '2':
            add_book()
        elif choice == '3':
            update_book()
        elif choice == '4':
            delete_book()
        elif choice == '5':
            print("Exiting menu.")
            break
        else:
            print("Invalid choice. Try again.")

menu()
```

```python
### 4 th Task: Movie Recommendation System
##Use this url http://www.omdbapi.com/ to fetch information about movies.
## Create a program that asks users for a movie genre and recommends a random movie from that genre.
import requests
import random
API_KEY = 'f8dd3113'  
OMDB_URL = 'http://www.omdbapi.com/'
def get_movies_by_genre(target_genre, movie_titles):
    matched_movies = []
    
    for keyword in movie_titles:
        params = {
            'apikey': API_KEY,
            's': keyword
        }
        response = requests.get(OMDB_URL, params=params)
        data = response.json()

        if response.status_code == 200 and data.get("Response") == "True":
            genres = [g.strip().lower() for g in data.get("Genre", "").split(',')]
            if target_genre.lower() in genres:
                matched_movies.append({
                    "title": data["Title"],
                    "year": data["Year"],
                    "genre": data["Genre"],
                    "plot": data["Plot"]
                })
    
    return matched_movies

sample_movies = [
    "Inception", "The Dark Knight", "Mad Max: Fury Road", "Avengers: Endgame",
    "Gladiator", "John Wick", "Die Hard", "The Matrix", "Black Panther",
    "Iron Man", "The Avengers", "Thor: Ragnarok", "Captain America: Civil War",
    "Tenet", "Mission: Impossible - Fallout"
]

def recommend_movie():
    genre = input("Enter a movie genre (e.g., Action, Drama, Comedy): ").strip()
    print(f"\nüîç Searching for {genre.title()} movies...")

    results = get_movies_by_genre(genre, sample_movies)
    
    if results:
        movie = random.choice(results)
        print("\nüé¨ Movie Recommendation:")
        print(f"Title : {movie['title']}")
        print(f"Year  : {movie['year']}")
        print(f"Genre : {movie['genre']}")
        print(f"Plot  : {movie['plot']}")
    else:
        print(f"‚ùå No movies found for genre: {genre}")

recommend_movie()
```
